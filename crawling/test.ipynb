```python
import ast

def transform_time(times_str):
    if pd.isnull(times_str):
        return times_str
    else:
        times_list = ast.literal_eval(times_str)
        
        new_times_list = []
        for time in times_list:
            day, times =  time.split(':', 1)
            # print(f'day: {day}, times: {times}')
            time_list = times.split('-')
            # print(time_list)
            if (len(times.split('-')) > 2):
                new_time = time_list[0] + '-' + time_list[-1]
                new_times_list.append(f'{day} : {new_time}')
                # print(new_time)
            else:
                return times_str
        return new_times_list
```
