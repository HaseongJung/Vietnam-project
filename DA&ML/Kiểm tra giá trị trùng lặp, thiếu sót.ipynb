```python
import pandas as pd
import nltk
import re
import numpy as np
```

```python
t_df = pd.read_csv('trip.csv')
t_df.drop(['Unnamed: 0'], axis = 1, inplace = True)
t_df
```

```python
dup_df = t_df.duplicated(subset=["name"])
print(dup_df)
```

```python
num_dup_df = dup_df.sum()
print(num_dup_df)
```

```python
# loại bỏ giá trị trùng lặp
drop_df = t_df.drop_duplicates(subset='name', keep='first').reset_index(drop=True)
print(len(drop_df))
```

```python
# Loại bỏ giá trị thiếu sót dựa trên tiêu chuẩn Category, Name, Reviews 
NaN_deleted = drop_df.dropna(subset=['category', 'name', 'reviews'], how='any', axis=0)

print(NaN_deleted.isnull().sum())
```
